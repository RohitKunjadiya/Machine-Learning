Density Based Spatial Clustering of Application with Noise

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
from sklearn.cluster import DBSCAN
```

```python
X = np.array([[1,2],[2,2],[2,3],[8,7],[8,8],[25,80]])
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1])
plt.show()
```

```python
db = DBSCAN(eps=3,min_samples=2)

db.fit(X)
db.labels_
```

```python
db = DBSCAN(eps=3,min_samples=3)

db.fit(X)
db.labels_
```

```python
db = DBSCAN(eps=10,min_samples=2)

db.fit(X)
db.labels_
```

```python
from sklearn.datasets import make_circles

X,_ =  make_circles(n_samples=1000,factor=0.5,noise=0.05,random_state=4)
```

```python
db = DBSCAN(eps=0.1,min_samples=5)

db.fit(X)
cluster = db.fit_predict(X)

sns.scatterplot(x=X[:,0],y=X[:,1],hue=cluster)
plt.title("DBSCAN Clustering of Concentric Circles")
plt.show()
```
