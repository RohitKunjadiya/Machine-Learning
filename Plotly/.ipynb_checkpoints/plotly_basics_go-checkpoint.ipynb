```python
import numpy as np
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
match = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

```python
ipl = delivery.merge(match , left_on = 'match_id' , right_on = 'id')
```

```python
ipl.head()
```

```python
# Scatter plots are drawn between to continous variables
# Problem :- We are going to draw a scatter plot between Batsman Avg(X axis) and
# Batsman Strike Rate(Y axis) of the top 50 batsman in IPL(All time)
```

```python
# Avg vs SR graph of Top 50 batsman(in terms of total runs)

# Fetching a new dataframe with Top 50 batsman

#top50 = list(ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).index)

top50 = ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(50).index.tolist()
new_ipl = ipl[ipl['batsman'].isin(top50)]
```

```python
#Calculating Strike Rate

runs = new_ipl.groupby('batsman')['batsman_runs'].sum()
balls = new_ipl.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
sr = sr.reset_index()
sr.rename(columns = {'batsman_runs':'sr'},inplace=True)
```

```python
#Calculating Average

# Calculating number of outs for top 50 batsman

out = ipl[ipl['player_dismissed'].isin(top50)]

outs = out['player_dismissed'].value_counts()

avg = runs/outs

avg = avg.reset_index()
avg.rename(columns = {'index':'batsman',0:'avg'},inplace=True)
```

```python
avg = avg.merge(sr,on='batsman')
```

```python
avg
```

# Scatter plot

```python
#Scater Plot

trace = go.Scatter(x=avg['avg'],y=avg['sr'],mode='markers',text=avg['batsman']
                  ,marker={'color':'#00a65a','size':19})

data=[trace]

layout = go.Layout(title = 'Batsman avg vs sr',xaxis={'title':'Batsman Average'},
                  yaxis={'title':'Batsman Strike Rate'})

fig = go.Figure(data=data,layout=layout)

pyo.plot(fig,filename='scatter.html')
#default name='temp-plot.html'
```

# Line Chart

it's an extension of Scatter plot. Usually used to show a time series data.

```python
#year-by-year batsman performance

single = ipl[ipl['batsman']=='V Kohli']
performance=single.groupby('season')['batsman_runs'].sum().reset_index()
performance
```

```python
trace = go.Scatter(x=performance['season'],
                  y=performance['batsman_runs'],
                  mode='lines',
                  marker={'color':'#00a65a'})

data=[trace]

layout = go.Layout(title = 'year-by-year performance',
                  xaxis={'title':'season'},
                  yaxis={'title':'Total runs'})

fig = go.Figure(data=data,layout=layout)

pyo.plot(fig,filename='lines.html')
```

```python
trace = go.Scatter(x=performance['season'],
                  y=performance['batsman_runs'],
                  mode='lines+markers',
                  marker={'color':'#00a65a'})

data=[trace]

layout = go.Layout(title = 'year-by-year performance',
                  xaxis={'title':'season'},
                  yaxis={'title':'Total runs'})

fig = go.Figure(data=data,layout=layout)

pyo.plot(fig,filename='lines1.html')
```

```python
#year-by-year batsman performance

single = ipl[ipl['batsman']=='V Kohli']
performance=single.groupby('season')['batsman_runs'].sum().reset_index()
performance

single1 = ipl[ipl['batsman']=='RG Sharma']
performance1 = single1.groupby('season')['batsman_runs'].sum().reset_index()
performance1
```

```python
trace = go.Scatter(x=performance['season'],
                  y=performance['batsman_runs'],
                  mode='lines+markers',
                  marker={'color':'#00a65a'},name='Virat Kohli')

trace1 = go.Scatter(x=performance1['season'],
                  y=performance1['batsman_runs'],
                  mode='lines+markers',name='Rohit Sharma')

data=[trace,trace1]

layout = go.Layout(title = 'year-by-year performance',
                  xaxis={'title':'season'},
                  yaxis={'title':'Total runs'})

fig = go.Figure(data=data,layout=layout)

pyo.plot(fig,filename='lines2.html')
```

```python
# Multiple Line Charts

def batsman_comp(*name):
    data=[]
    for i in name:
        single=ipl[ipl['batsman']==i]
        performance=single.groupby('season')['batsman_runs'].sum().reset_index()
        trace=go.Scatter(x=performance['season'],y=performance['batsman_runs']
                         ,mode='lines + markers',name=i)
        
        data.append(trace)
    
    layout=go.Layout(title='Batsman Record Comparator',
                xaxis={'title':'Season'},
                yaxis={'title':'Runs'})

    fig=go.Figure(data=data,layout=layout)

    pyo.plot(fig,filename='year_by_year.html')
```

```python
batsman_comp('V Kohli', 'RG Sharma','DA Warner','MS Dhoni')
```

# Bar Plot

Used to show relation between one categorical and 1 numerical data

```python
top10 = ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10).index.tolist()
top = ipl[ipl['batsman'].isin(top10)]
```

```python
score = top.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
#plot the bar graph

trace = go.Bar(x=score['batsman'],y=score['batsman_runs'])

data = [trace]

layout = go.Layout(title = 'Top 10 IPL Batsman',
                  xaxis={'title':'batsman'},
                  yaxis={'title':'Total runs'})

fig = go.Figure(data=data,layout=layout)
pyo.plot(fig,filename='bar.html')
```

# There are 3 types of Bar Graphs
1. Nested Bar Graph
2. Stacked Bar Graph
3. Overlayed Bar Graph

```python
iw = top.groupby(['batsman','inning'])['batsman_runs'].sum().reset_index()
mask = iw['inning']==1
mask2 = iw['inning']==2
one = iw[mask]
two = iw[mask2]


one.rename(columns={'batsman_runs':'1st Innings'},inplace=True)
two.rename(columns={'batsman_runs':'2nd Innings'},inplace=True)

final=one.merge(two,on='batsman')[['batsman','1st Innings','2nd Innings']]

final
```

# Overlayed Bar Graph

```python
trace = go.Bar(x=final['batsman'],
                  y=final['1st Innings'],
                  marker={'color':'#00a65a'})

trace1 = go.Bar(x=final['batsman'],
                  y=final['2nd Innings'])

data=[trace,trace1]

layout = go.Layout(title = 'Inning Wise Score',
                  xaxis={'title':'Batsman'},
                  yaxis={'title':'Btasman runs'},
                  barmode='overlay')

fig = go.Figure(data=data,layout=layout)

pyo.plot(fig,filename='inning_comarision_overlayed.html')
```

# Satcked Bar Graph

```python
trace = go.Bar(x=final['batsman'],
                  y=final['1st Innings'],
                  marker={'color':'#00a65a'})

trace1 = go.Bar(x=final['batsman'],
                  y=final['2nd Innings'])

data=[trace,trace1]

layout = go.Layout(title = 'Inning Wise Score',
                  xaxis={'title':'Batsman'},
                  yaxis={'title':'Btasman runs'},
                  barmode='stack')

fig = go.Figure(data=data,layout=layout)

pyo.plot(fig,filename='inning_comarision_stacked.html')
```

# Nested Bar Graph

```python
trace = go.Bar(x=final['batsman'],
                  y=final['1st Innings'],
                  marker={'color':'#00a65a'})

trace1 = go.Bar(x=final['batsman'],
                  y=final['2nd Innings'])

data=[trace,trace1]

layout = go.Layout(title = 'Inning Wise Score',
                  xaxis={'title':'Batsman'},
                  yaxis={'title':'Btasman runs'})

fig = go.Figure(data=data,layout=layout)

pyo.plot(fig,filename='inning_comarision_nested.html')
```

# Box Plot

A box and whisker plotâ€”also called a box plotâ€”displays the five-number summary of a set of data.

```python
match_agg=delivery.groupby(['match_id'])['total_runs'].sum().reset_index()
season_wise=match_agg.merge(match,left_on='match_id',right_on='id')[['match_id','total_runs','season']]
season_wise
```

```python
match_agg
```

```python
trace = go.Box(x=season_wise['season'],name='All seasons')

data = [trace]

layout = go.Layout(title='total score analysis',
                  xaxis={'title':'Total score'})

fig = go.Figure(data=data,layout=layout)

pyo.plot(fig,filename='runs_in_match_box.html')
```

```python
trace1 = go.Box(x=season_wise[season_wise['season']==2008]['total_runs'],name='2008')

trace2 = go.Box(x=season_wise[season_wise['season']==2017]['total_runs'],name='2017')

data = [trace1,trace2]

layout = go.Layout(title='total score analysis',
                  xaxis={'title':'Total score'})

fig = go.Figure(data=data,layout=layout)

pyo.plot(fig,filename='2008_2017_runs_in_match_box.html')
```

# Histograms

A histogram is a plot that lets you discover, and show, the underlying frequency distribution (shape) of a set of continuous data.

```python
x=delivery.groupby('batsman')['batsman_runs'].count()>150
x=x[x].index.tolist()

new=delivery[delivery['batsman'].isin(x)]


runs=new.groupby('batsman')['batsman_runs'].sum()
balls=new.groupby('batsman')['batsman_runs'].count()

sr=(runs/balls)*100

sr=sr.reset_index()
sr.rename(columns={'batsman_runs':'sr'},inplace=True)
sr
```

```python
trace = go.Histogram(x=sr['sr'],name='Strike Rate',xbins={'size':5,'start':80,'end':120})

data = [trace]

layout = go.Layout(title='Strike Rate analysis',
                  xaxis={'title':'Strike Rates'})

fig = go.Figure(data=data,layout=layout)

pyo.plot(fig,filename='histogram.html')
```

# Heatmaps

A heat map is a graphical representation of data where the individual values contained in a matrix are represented as colors.

```python
six=delivery[delivery['batsman_runs']==6]
six=six.groupby(['batting_team','over'])['batsman_runs'].count().reset_index()

six.rename(columns={'batsman_runs':'six'},inplace=True)
six
```

```python
trace = go.Heatmap(x=six['batting_team'],y=six['over'],z=six['six'])

data = [trace]

layout = go.Layout(title='six heatmap')

fig = go.Figure(data=data,layout=layout)

pyo.plot(fig,filename='heatmap.html')
```

```python
# Side by Side Heatmap

dots=delivery[delivery['batsman_runs']==0]
dots=dots.groupby(['batting_team','over'])['batsman_runs'].count().reset_index()

from plotly import tools

trace1=go.Heatmap(x=six['batting_team'],y=six['over'],
                 z=six['six'].values.tolist())

trace2=go.Heatmap(x=dots['batting_team'],y=dots['over'],
                 z=dots['batsman_runs'].values.tolist())


fig=tools.make_subplots(rows=1,cols=2,subplot_titles=["6's","0's"], shared_yaxes=True)

fig.append_trace(trace1,1,1)
fig.append_trace(trace2,1,2)

pyo.plot(fig,filename='heatmap1.html')
```
